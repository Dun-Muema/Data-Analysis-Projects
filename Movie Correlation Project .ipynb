```python
# import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the confirguration of the plots I will create

# Read the Data 

df = pd.read_csv(r'C:\Users\DuncanMu\Documents\Projects\movies.csv')
```

```python
# inspect the Data

df.head()
```

```python
# Check for any Missing Data

for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
# Inspect Data types for all the columns

df.dtypes
```

```python
# Replace null values in Budget and Gross

df['budget'].fillna(0, inplace=True)
df['gross'].fillna(0, inplace=True)
```

```python
# change data types for columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
#Getting the correct movie release year from the released date 

df['year_correct'] =df['released'].astype(str).str[:4]

df.head()
```

```python
# Sorting the Movies by Gross profit in Descending order

df = df.sort_values(by=['gross'],inplace= False, ascending= False)
```

```python
# Displaying all the rows in the dataset

pd.set_option('display.max_rows', None)
```

```python
# Drop any Duplicates

df['company'].drop_duplicates().sort_values(ascending =False)
```

```python
df.head()
```

```python
# scatter plot with budget vs gross (Check the Correlation)

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
# plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color": "green"})
```

```python
correlation_matrix = df.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot =True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# inpect the column(company)

df.head()
```

```python
#convert the Column dtypes from Object to category for easy use in a correlation matrix

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot =True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
